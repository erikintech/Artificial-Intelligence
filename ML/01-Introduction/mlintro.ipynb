## **History of AI**

- 1950: Alan Turing developed the Turing test, to test a machine's ability to exhibit intelligent behavior.
- 1956: Artificial Intelligence was accepted as a field at the Dartmouth Conference
- 1957: Frank Rosenblatt invented the perceptron algorithm, this was the precursor to modern neural networks.
- 1959: Arthur Samuel published an algorithm for a checkers program using machine learning

### **AI Winter**
The AI Winter is the period of reduced funding and interest in artificial intelligence research.

### **The First AI Winter**
- 1966: ALPAC committeeevaluated AI techniques for machine translation and determined there was little yield from the investment
- 1969:Marvin Minky published a book on the limitations of the Perceptron algortihm wich slowed research in neural networks
- 1973: The Lighthill report highlights AI's failure to live up to promise.

### **1980's AI Boom**
**Expert Systems** - systems with programmed rules designed to mimic human experts
- Ran on mainframe computers with specialized programming languages like LISP
- Were the first widely-used AI technoly with two-thirds of Fortyme 500 companies using theme at their peek
- 1986: The "Backpropagation" algorithm is able to train multi-layers perceptrons, leading to new successes and interest in neural networks.

### **Another AI Winter (1980's - 1990's)**

Expert systems progress on solving business problems slowed.
- Expert systems began to be converted into software suites of general business applicationslike SAP or Oracle, that could run on PC's instead of mainframes.
- Neural networks didn't scale to large problems.
- Interest of AI in business declined.

### **1990s - 2000s Maachine Learning**

AI solutions gad successes in speech recognition, medical diagnosis, robotics and other areas.
- AI algorithms were integrated into larger systems and became useful throughout industry.
- The Deep Blue chess system beat world chess champion Garry Kasparov
- Google's search engine launched using artificial intelligence technology.

### **2006 Rise of Deep Learning**
- 2006 Geoffrey Hinton publishes a paper on unsupervised pre-training that allowed deeper neural networks to be trained.
   - Neural networks are rebranded as deep learning
- 2009 The ImageNet database of human-tagged images is presented at the CVPR conference.
- 2010 Algorithms compete on several visual recognition tasks at the first ImageNet competition.

### **Deep Learning Breakthroughs (2012 - ?)**
- 2012 Deep learning beats previous benchmark on the ImageNet cmpetition.
- 2013 Deep learning is used to understand "conceptual meaning" of words
- 2014 Similar breakthroughs appeared in language translation.

These have led to advancements in Web Search, Document Search, Document Summarization and Machine Translation

- 2014 Stanford team creates computer vision algorithm that can describe photos
- 2015 Deep learning platform TensorFlow is developed by Google
2016 DeepMind's AlphaGo, developed by Aja Huang beats Go master Lee Se-dol
- 2018 Waymo lauches commercialself-driving-car service in suburbs of Phoenix
- 2019 IBM Project Debater is able to have a full debate with rebuttal with champion human debater

### **Machine Learning Algorithm**

## **Supervised Learning**

- This type of algorithms build a mathematical model
- Contains inputs and outputs
- Used historical and labeled data, then the ML model **predicts** a value
- Knows results and data from the past
- Labeled data means that we know the features of the data and past prediction labels
- Have a training data set and a test data set

#### **Two Tupes of Supervised Algorithms**
- Regression - continuos value to predict
   - e.g. predict the price of a product, predict the electricity loads, etc
- Classification - categorical value to predict.
   - e.g. handwriting detection, spam filtering, fake news detection, etc.
   
#### **Algorithms**
- Linear Regression
- Linear Regression + Regularization
- Polynomial Regression
- Support Vector Machine
- K-Nearest Neighbors
- Decision Trees
- Random Forest
- XGBoost
- Gradient Decent Boosting
- CatBoost
- Baive Bayes

#### **Applications**
- Handwriting detection
- Bioinformatics
- House price prediction
- Sales forecasting
- Image classification

## **Unsupervised Learning**
- Applied to unlabeled  data, the ML model discovers possible patterns in the data
- Desired output is unknow
- Trying to group and label data without alabel.
- Downside is that there is no historical correct labe, it's hard to evaluate performance.

#### **Algorithms**
- K Means
- K Prototypes
- DBSCAN
- Hierarchical
- PCA
#### **Applications**
- Customer SEgmentation
- Clustering
- Dimensionality reduction
- Finding association rules
