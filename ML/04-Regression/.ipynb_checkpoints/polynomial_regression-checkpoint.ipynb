# Polynomial Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
X = df.drop('Sales', axis=1)
```

```python
X.head()
```

```python
y = df.Sales
```

```python
y.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_converter = PolynomialFeatures(degree= 2, include_bias= False)
```

```python
# Converting X features in polynomial features
poly_converter.fit(X)
```

```python
poly_converter.transform(X)
```

```python
poly_features = poly_converter.transform(X)
```

```python
poly_features.shape
```

```python
X.iloc[0]
```

```python
poly_features[0]
```

#### Train model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
# training
model.fit(X_train, y_train)
```

```python
# coeff
model.coef_
```

```python
# intercept value
model.intercept_
```

```python
test_predicition = model.predict(X_test)
```

```python
test_predicition
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, test_predicition)
```

```python
MSE = mean_squared_error(y_test, test_predicition)
```

```python
RMSE = np.sqrt(MSE)
```

```python
print(f'MAE: {MAE}\nMSE: {MSE}\nRMSE: {RMSE}')
```

#### Choosing the polunomial degree

```python
# Create the different order poly
# split poly feat train/test
# fit on train
#store/save the rmse for both the train and test
# plot the results error vs poly order
```

```python
train_rmse_errors = []
test_rmse_errors = []

for d in range(1, 10):
    poly_converter = PolynomialFeatures(degree=d, include_bias=False)
    poly_features = poly_converter.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    
    train_rmse = np.sqrt(mean_squared_error(y_train, train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, test_pred))
    
    train_rmse_errors.append(train_rmse)
    test_rmse_errors.append(test_rmse)
```

```python
train_rmse_errors
```

```python
test_rmse_errors
```

```python
plt.plot(range(1, 6), train_rmse_errors[:5], label='TRAIN RMSE', color='white')
plt.plot(range(1, 6), test_rmse_errors[:5], label='TEST RMSE', color='red')
plt.ylabel('RMSE')
plt.xlabel('Poly Degree')
plt.legend()
```

```python
final_poly_converter = PolynomialFeatures(degree=3, include_bias=False)
```

```python
final_model = LinearRegression()
```

```python
full_converted = final_poly_converter.fit_transform(X)
final_model.fit(full_converted, y)
```

```python
from joblib import dump, load
```

```python
dump(final_model, 'final_poly_model.joblib')
```

```python
dump(final_poly_converter, 'final_poly_converted.joblib')
```
