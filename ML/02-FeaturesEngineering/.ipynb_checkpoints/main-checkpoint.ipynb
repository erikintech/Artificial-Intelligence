# <div align=center> **Features Engineering** </div>

## **Definition**
Feature engineering refers to manipulation, addition, deletion, combination, mutation of a data set to improve machine learning/ deep learningmodeltraining, leading to better performance andgreater accuracy

Some process of features engineer should be done after the train/test split process, so here you can a relative:

* Handling missing values
* Handling outliers
* **Train/Test Split**
* Scaling | Normalize data -> only numerical data
* OneHotEncoding -> categorical predictors
* LabelEncoder (if it's classification problem) -> applied to the target
* Feature selection

*The features engineering process will depend on the type of ml algorithms and the type of data that we are using.*

## **1. Handling Missing Values**

In some data set we can find NA values, it is not nothing but missing data. We can handling this type of data of different ways:

- Replace missing values with mean or median for nuerical data and mode for categorical data.
- Drop NA values entire rows
- Drop columns with NA values
- Replace NA values with 0

#### **When can you use median or mean to replace NA vaalues?**
- Mean when the data is symmetric
- Median when the data is skewed

````python
# getting data
df = pd.read_csv('../../datasets/melb_data.csv')

# getting shape of data set
df.shape
````

````python
# getting info about columns
df.info()
````

````python
# Vrifying for missing values
df.isna().sum()
````

In this case we have 4 features with missing values, Car, bBuildingArea and YearBuilt are numerical dta, so we can use medain or mean to replace missing values, then we have CouncilArea feature that is an object data type, here we can use mode

````python
# plot numerical features

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))
sns.histplot( ax=axes[0], data=df, x='Car', bins=25 )
sns.histplot( ax=axes[1], data=df, x='BuildingArea', bins=25 )
sns.histplot( ax=axes[2], data=df, x='YearBuilt', bins=25 )

fig.suptitle('Numerical Features Distribution Type')
````

In this case all numerical data with NA values are skewed, so I will replace missing values with median of each feature

````python
# refill missing values -> numerical features
for feature in ['Car', 'BuildingArea', 'YearBuilt']:
    df[feature] = df[feature].fillna( df[feature].median() )
    
# categorical feature
df['CouncilArea'] = df['CouncilArea'].fillna( df['CouncilArea'].mode()[0] )
````

## **2. Handling Outliers**

<img src='./outliers.png'/>

**Outliers** are data points that difeer significatly from other observation.

**How can e identify outliers?**
- Firstly, calculate the skewness of features and check wether they are positively skewed, negatively skewed or normally skewed. Another method is plot a boxplot to features and check if there are values out of extremes, if there, they are called outliers

**How to handle outliers?**
0. Identify outliers
1. Calculate the quantile values at 25% and 75%
2. Calculate the interquartile value
3. Calculate the upper extreme and lower extreme
4. Chceck the values will lie above the upper extreme or belowe the lower extreme, if it presents then remove them or replace them with the median, mean or any quantile value.

```python
# Replace outlier with quantile values
outliers = df[(df['Price']>upper_extreme)]['Price'].values
df['Price'].replace( outliers, upper_extreme, inplace=True )

# Replace outlier with mean
mean = np.around(df['Price'].mean(), 2)
df['Price'].replace( outliers, mean, inplace=True )

# Drop outliers
out = df[ (df['Price'] > upper_extreme) | (df['Price'] < lower_extreme) ].index
df.drop(out, inplace=True)
```

````python
numerical_columns = list(df.select_dtypes(include=['float64', 'int64']).columns)

# dropping outliers
for feature in numerical_columns:
    q1 = df[feature].quantile(0.25)
    q3 = df[feature].quantile(0.75)
    iqr = q3 - q1
    lower_extreme = q1 - 1.5*iqr
    upper_extreme = q3 + 1.5*iqr
    outliers = df[ (df[feature] < lower_extreme) | (df[feature]>upper_extreme) ].index
    df.drop(outliers, inplace=True)
````

## **3. Train | Test Split**

This step consists in split our whole data set in a train and test sets, in general we use between 70%-80% for train data set and the 20%30% for test set

````python
df.info()
````

````python
from sklearn.model_selection import train_test_split

# split in features and target
X = df.drop(['Price', 'Suburb', 'Address', 'SellerG', 'Date'], axis=1)
y = df['Price']

# split in train and test sets
# 80% Train - 20% Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
````

## **4. Encoding**

In a dataset we can find object datatype and numerical datatypes likefloat or integer, machine learning models are good friend with numerical data, so if we want a better performance in our ML model we need to transform categorical data into numerical data. We have two method OneHotEncoding for features and LabelEncoding for target variable.

````python
# import OneHotEncoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_data = list( X_train.select_dtypes('object').columns )

ct = ColumnTransformer(
    [ ('encoder', OneHotEncoder(handle_unknown='ignore',sparse_output=False ), categorical_data ) ],
    remainder='passthrough', verbose_feature_names_out=False
)

encoded_X_train = pd.DataFrame( ct.fit_transform(X_train), columns= ct.get_feature_names_out() )
encoded_X_test = pd.DataFrame( ct.transform(X_test), columns= ct.get_feature_names_out() )
````

## **5. Feature Scaling**

## **Why should we use feature scaling?**

Some ML algorithms are sensitive to feature scaling, and others are invariant to it.

**Gradient Descent Based Algorithms**

Algorithms like linear regression, logistic regression that use gradient descent as an optimization technique require data to be scaled.

**Distance Based Algorithm**

Algorithms like KNN, Support Vector MAchine use distances between data point to determine their similarity, for this reason when we use this type of algorithms is important that we use feature scaling

## **Normalization**

Normalization is a technique for scaling data in a range between 0 and 1, this is also know as **Min-Max Scaling**

***Formula:***
$$
 \tilde X= \frac{X - X_{min}}{X_{max} - X_{min} }
$$

- When the value of X is the minimun in the columns, the numerator will be 0 and $\tilde X$ is 0
- When X is the maximun value in  the column , the numerator is equal to the denominator and thus the value of $\tilde X$ is 1

## **Standarization**

It is the other technique of scaling where the values are centered around the mean with a unit standard deviation.

***Formula:***

$$
    \grave X = \frac{X - \nu }{\alpha}
$$

*Where:*

- **$\nu$** represents the mean of the feature values

- **$\alpha$** represents the standard deviation of te feature values

### **When should we use Normalization or Standarization?**

- **Normalization** is used when you know that the distribution of your data does not follow a Gaussian distribution, in algorithms like K-Nearest Neighbors and Neural Networks.
- **Standarization** is used when you know that your data follows a Gaussian distribution, standarization can help us specially when our data have outliers

This step is applied only in the features for both sets, train and test

````python
# applying mixmaxscaler
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Get numerical features
numerical_features = list(encoded_X_train.select_dtypes(['int64', 'float64']).columns)

scaler_ct = ColumnTransformer(
    [ ( 'scaler', MinMaxScaler(), numerical_features ) ],
    remainder='passthrough', verbose_feature_names_out=False
)
scaled_X_train = pd.DataFrame( scaler_ct.fit_transform(encoded_X_train), columns=scaler_ct.get_feature_names_out() )
scaled_X_test = pd.DataFrame( scaler_ct.transform(encoded_X_train), columns=scaler_ct.get_feature_names_out() )
````

## **6. Feature Selection**

Feature selection methods are intended to reduce the number of predictrs to those that are believed to be most useful to a model in order to predict a target

We have two types of data
- **Numerical**
   * Float
   * Integer
* **Categorical**
   * Ordinal
   * Boolean
   * Nominal

### **How to choose a feature selection method?**

Well we have two types of imput data numerical input and categorical input, for each data input type we can have numerical input or categorical input, so we have the following:

**Numerical input, numerical output**
This is a regression problem so in this case we can use correlation coefficient, or rank-based methods for a nonlinear data.
* Pearson's correlation coefficient (linear data)
* Spearman's rank coefficient (nonlinear data)

**Numerical input, Categorical output**

Classification problem
* ANOVA correlation coefficient (linear data)
* Kendall rank coefficient (nonlinear)
*Kendall  does assume that the categorical data is ordinal, so we have to encode data to do feature selection in both cases ANOVA and kendall*

**Categorical input, Numerical output**

This a strange case of regression problem, you can use the same methods of *Numerical input, Categorical output*

**Categorical input, Categorical output**

Classification problem

* Chi-Squared coefficient
* Mutual information

#### **Benefits of selecting best features**

- Reduces overfitting
- Improves accuracy
- Reduces training time

````python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression # Pearson's correlation

selector = SelectKBest(score_func=f_regression, k=20) # the best 20 features
selected_X_train = pd.DataFrame( selector.fit_transform(encoded_X_train, y_train), 
                                columns=selector.get_feature_names_out() )
selected_X_test = pd.DataFrame( selector.transform(encoded_X_test),
                              columns=selector.get_feature_names_out())
````
